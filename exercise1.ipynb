Task (Grace units): The manager decides to add **2 "bonus" units** to every products daily sale for a holiday report. add  2 to the entire daily_sales array. task2.2 (Revenue): Calculate the total revenue for each day.

Multiply the cost_price array by the daily_sales matrix.
Observe how numpy handles the multiplication of  1d and 2 array.

Task2.3 (Scaling): Use np.sqrt() on the sales data to find the square root of all sales values for statistical curve.

```python
import numpy as np

# daily sales of 3 products over 5 days
daily_sales = np.array([
    [10, 12, 8, 9, 11],
    [5, 7, 6, 8, 7],
    [20, 18, 22, 19, 21]
])

# cost price of each product
cost_price = np.array([100, 60, 150])
```

```python
daily_sales_bonus = daily_sales + 2
print(daily_sales_bonus)
```

```python
revenue = daily_sales_bonus * cost_price[:, np.newaxis]
print(revenue)
```

```python
total_revenue_per_day = revenue.sum(axis=0)
print(total_revenue_per_day)
```

```python
sqrt_sales = np.sqrt(daily_sales_bonus)
print(sqrt_sales)
```

# Where in Numpy

```python
arr = np.array([10, -5, 20, -3, 15])
result = np.where(arr < 0, 0, arr)
print(result)
```

```python
arr = np.where(arr % 2!=0, -1, arr)
print(result)
```

```python
marks= np.array([45, 78, 32, 90, 55])
status = np.where(marks>=50,1,0)

print(status)
```

```python
data=np.array([10,20,30,40,50])
mean_val = np.mean(data)
result=np.where(data>mean_val, mean_val,data)
```
